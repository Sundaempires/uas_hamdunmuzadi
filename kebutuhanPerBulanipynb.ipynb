```python
# Membaca file CSV
data = pd.read_csv('/content/kebutuhan_per_bulan_2.csv')

# Mengambil kolom yang dibutuhkan
data_prov = data['Provinsi']
data_thn = data['tahun']
data_kbthn = data['kebutuhan']

# Membuat scatter plot dengan ukuran gambar yang diperlebar
plt.figure(figsize=(12, 6))

# Menentukan warna berdasarkan data_kebutuhan
colors = cm.get_cmap('viridis', len(data_kbthn))
color_mapping = {value: colors(i) for i, value in enumerate(data_kbthn)}

# Membuat scatter plot dengan warna berdasarkan data_kebutuhan
plt.scatter(data_prov, data_thn, c=[color_mapping[value] for value in data_kbthn], alpha=0.7)

# Mengatur judul dan label sumbu
plt.title('Scatter Plot - Kebutuhan per Tahun')
plt.xlabel('Provinsi')
plt.ylabel('Tahun')

# Memiringkan label sumbu x
plt.xticks(rotation=90)

# Menampilkan scatter plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Membaca file CSV
data = pd.read_csv('/content/kebutuhan_per_bulan_2.csv')

# Mengambil kolom yang dibutuhkan
data_prov = data['Provinsi']
data_thn = data['tahun']
data_kbthn = data['kebutuhan']

# Mengatur warna berdasarkan data_kbthn
colors = plt.cm.viridis(np.linspace(0, 1, len(data_kbthn)))

# Membuat scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(data_prov, data_thn, c=colors, alpha=0.7)

# Menambahkan colorbar sebagai keterangan warna
sm = plt.cm.ScalarMappable(cmap=plt.cm.viridis)
sm.set_array([])
plt.colorbar(sm, label='Kebutuhan')

# Mengatur judul dan label sumbu
plt.title('Scatter Plot - Kebutuhan per Tahun')
plt.xlabel('Provinsi')
plt.ylabel('Tahun')

# Memiringkan label sumbu x
plt.xticks(rotation=90)

# Menampilkan scatter plot
plt.show()
```
