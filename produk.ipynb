```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv('/content/produk.csv')

data_kode = data['Kode']
data_harga = data['Harga']
data_produk = data['Produk']

label_encoder = LabelEncoder()
data_produk_encoded = label_encoder.fit_transform(data_produk)

plt.scatter(data_kode, data_harga, c=data_produk_encoded, cmap='viridis')

plt.colorbar(label='Produk')

plt.xlabel('Kode')
plt.ylabel('Harga')

for i, produk in enumerate(data_produk):
    plt.annotate(produk, (data_kode[i], data_harga[i]), textcoords="offset points",
                 xytext=(5, 5), ha='left', fontsize=8)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv('/content/produk.csv')

data_kode = data['Kode']
data_harga = data['Harga']
data_produk = data['Produk']

label_encoder = LabelEncoder()
data_produk_encoded = label_encoder.fit_transform(data_produk)

plt.scatter(data_kode, data_harga, c=data_produk_encoded, cmap='viridis')

cbar = plt.colorbar(label='Produk')

cbar.set_ticks(range(len(label_encoder.classes_)))

cbar.set_ticklabels(label_encoder.classes_)

plt.xlabel('Kode')
plt.ylabel('Harga')

plt.show()
```
