```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/kebutuhan_per_bulan_2.csv')

plt.figure(figsize=(12, 6))
plt.scatter(data['Propinsi'], data['tahun'], s=50, c='blue', alpha=0.7)

for i in range(len(data)):
    plt.annotate(data['kebutuhan'][i], (data['Propinsi'][i], data['tahun'][i]), ha='center', va='bottom', fontsize=8)

plt.title('Kebutuhan Per Tahun Berdasarkan Provinsi', fontsize=14)
plt.xlabel('Propinsi', fontsize=12)
plt.ylabel('Tahun', fontsize=12)

plt.xticks(rotation=90, fontsize=8)

plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/kebutuhan_per_bulan_2.csv')

data_filtered_jabar = data[data['Propinsi'] == 'JAWA BARAT']

plt.scatter(data_filtered_jabar['tahun'], data_filtered_jabar['kebutuhan'], c=data_filtered_jabar.index, cmap='rainbow', s=50, alpha=0.7)

plt.title('Scatter Plot - Kebutuhan per Tahun (Jawa Barat)')
plt.xlabel('Tahun')
plt.ylabel('Kebutuhan')

cbar = plt.colorbar()
cbar.set_label('Data Index')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/kebutuhan_per_bulan_2.csv')

fig, ax1 = plt.subplots(figsize=(12, 6))
ax1.scatter(data['Propinsi'], data['tahun'], s=50, c='red', alpha=0.7)
ax1.set_xlabel('Propinsi', fontsize=12)
ax1.set_ylabel('kebutuhan', color='red', fontsize=12)

for i in range(len(data)):
    ax1.annotate(data['tahun'][i], (data['Propinsi'][i], data['tahun'][i]), ha='center', va='bottom', fontsize=8)

ax2 = ax1.twinx()
ax2.bar(data['Propinsi'], data['kebutuhan'], color='blue', alpha=0.7)
ax2.set_ylabel('Kebutuhan', color='blue', fontsize=12)

ax1.set_xticklabels(data['Propinsi'], rotation=90, fontsize=8)

plt.title('Kebutuhan dan Tahun Berdasarkan Provinsi', fontsize=14)

ax1.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
